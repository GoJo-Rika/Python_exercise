## 10.1 Graphical representation of functions

The basic instruction for graphically representing a data collection is plt.plot(). This function graphically represents the data entered as an argument.

In the mathematical representation of a function or a collection of experimental data, we usually have the values ​​of the abscissa and the corresponding values ​​of the ordinate.

If we introduce two Numpy vectors as arguments of the `plt.plot ()` function then the values ​​of the first will correspond to the abscissa (_x_ axis) and those of the second to the ordinate (_y_ axis):

```python
plt.plot (x_data, y_data)
```

As a first experience with Matplotlib we will generate, step by step, the graphical representation of the function:

$$ f (x) = x ^ 2 $$

- We will write the instruction `% matplotlib inline` to ensure that the graph can be seen in the browser.
- We will add the `import` instructions for` numpy` and `matplotlib`.
- Using the `numpy.linspace ()` function we will generate an array containing 101 values ​​spaced between -2.0 and 2.0, which we will call "` x` ".
- We will generate a new array called "` y` "that contains the value of the $ x ^ 2 $ function at all points in the previous` x` array.
- Using the instruction `plt.plot ()` we will plot `y` as a function of` x`.
- We will add the `plt.show` statement so that the graph is displayed (this statement must go to the end of all` plt`).

In this way it is:
```Python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace (-2,2,101)
y = x ** 2
plt.plot (x, y) # Vectors x, y
plt.show ()
```
Check it in the cell below.

````python
%matplotlib inline

x = np.linspace (-2,2,101)
y = x ** 2
# print(x)

plt.plot (x, y) # Vectors x, y
plt.title("Graph of x^2 vs x")
plt.show ()
````

````python
%matplotlib inline

x = np.linspace(0, 3*np.pi, 500)

plt.plot(x, np.sin(x**2))
plt.title('A simple chirp')
````

Note that if only the Numpy vector corresponding to the ordinate (the _y_) is placed in the plot statement, then the indices of the elements are represented on the abscissa (in this case 0, 1, ... , 100).

Check it out by running it now:


### 10.1.1 Adding titles to axes

We can improve the presentation of the chart by adding titles to the axes using the instructions:

- plt.xlabel ("x_extext")
- plt.ylabel ("y_text")

These instructions can be set before or after the `plt.plot` statement but must go before the` plt.show () `statement.

Check how it works by copying and pasting the new version of the program we made.
```Python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y) # Vectors x, y
plt.show ()
```

````python
%matplotlib inline

x = np.linspace (-2,2,101)
y = x ** 2

plt.xlabel ("x")
plt.ylabel ("f (x)")
plt.plot (x, y) # Vectors x, y
plt.show ()
````

### 10.1.2 Add title to chart

We can also add a title to the chart with the instruction:

- plt.title ("chart_title")

Repeat the chart in the next cell, adding the title _"Graphical representation of x\**2"_

````python
%matplotlib inline

x = np.linspace (-2,2,101)
y = x ** 2

plt.xlabel ("x")
plt.ylabel ("f (x)")
plt.title("Chart Title")
plt.plot (x, y) # Vectors x, y
plt.show ()
````

### 10.1.3 Vary the representation limits

The `plt.plot ()` statement automatically adjusts the axis representation limit. It may be the case that we want to manually change these limits to highlight some feature of the function we are representing, for example highlight a maximum or expand a certain area. If so, we need to use the following features:

- `plt.xlim (xmin, xmax)`: Redefines the x-axis boundaries between the xmin and xmax values (previously defined).
- `plt.ylim (ymin, ymax)`: Redefines the boundaries of the y-axis between the ymin and ymax values (previously defined).

Copy the previous program into the next cell and modify it to represent the function in the $ x \ in [-1.5, 1.5] $ and $ y \ in [-0.5, 2.5] $ window.

````python
%matplotlib inline

x = np.linspace (-2,2,101)
y = x ** 2

plt.xlabel ("x")
plt.ylabel ("f (x)")
plt.xlim(-1.5,1.5)
plt.ylim(-0.5, 2.5)
plt.title("Chart Title")
plt.plot (x, y) # Vectors x, y
plt.show ()
````

### 10.1.4 Represent several functions simultaneously

If we want to graphically represent two functions in the same graph, we only need to write the `plt.plot ()` statement twice, one for each function we want to represent.

So for example, if we want to represent the functions $ x ^ 2 $ and $ x ^ 4 $ in the same graph, we can write:
```Python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y) # Vectors x, y
y2 = x ** 4
plt.plot (x, y2) # Vectors x, y2
plt.show ()
```
Copy and paste this code into the cell below and run it.

````python
%matplotlib inline

x = np.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y) # Vectors x, y
y2 = x ** 4
plt.plot (x, y2) # Vectors x, y2
plt.show ()
````

### 10.1.5 Legend

When we represent different curves simultaneously in the same figure, it is convenient to represent them so that they can be differentiated.

Within the `plt.plot ()` statement, a label can be assigned to the represented data, which is why the keyword `label` must be used in the form` label = "text_label" `. If we have more than one curve to represent, we can put a different label for each of them.

After the `plt.plot ()` statements (one for each curve) we need to type the `plt.legend ()` statement in order for a window to appear in the graph, inside a box called a legend, where 'will write the different tags.

Copy and paste the following code and check.
```Python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y, label = "x ** 2") # Vectors x, y
y2 = x ** 4
plt.plot (x, y2, label = "x ** 4") # Vectors x, y2
plt.legend ()
plt.show ()
```

````python
%matplotlib inline

x = np.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y, label = "x ^ 2") # Vectors x, y
y2 = x ** 4
plt.plot (x, y2, label = "x ^ 4") # Vectors x, y2
plt.legend ()
plt.show ()
````

### 10.1.6 Representation Styles

In order to be able to differentiate even more different curves made within the same graph, there is also the possibility of indicating for each of them the **style of representation**.

This is by default **a solid line**, but can be changed by line to strokes, dots, symbols, etc. In addition, you can indicate the color with which you want the curve to be represented.

The rendering style should be put in the `plt.plot` statement after the data to be rendered.

As an example, copy and paste the following code and see how the curves represent:
```Python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace(-2,2,11)
plt.xlabel("x")
plt.ylabel("f (x)")
y2 = x**2
plt.plot(x, y2, 'g', label = "x**2") # Green solid line (g: green)
y3 = x**3
plt.plot(x, y3, 'ro', label = "x**3") # Red dot (r: red) of large dots (o)
y4 = x**4
plt.plot(x, y4, 'b.', label = "x**4") # Small dotted line (b: blue) (.)
plt.legend()
plt.show()
```

````python
%matplotlib inline

x = np.linspace(-2,2,11)
plt.xlabel("x")
plt.ylabel("f (x)")
y2 = x**2
plt.plot(x, y2, 'g', label = "x**2") # Green solid line (g: green)
y3 = x**3
plt.plot(x, y3, 'ro', label = "x**3") # Red dot (r: red) of large dots (o)
y4 = x**4
plt.plot(x, y4, 'b.', label = "x**4") # Small dotted line (b: blue) (.)
plt.legend()
plt.show()
````

For more rendering style options you can consult **the package documentation [Matplotlib](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)** .

## 10.2 Generating files with graphics

The matplotlib module also allows you to save graphics to files. The function that allows it is `plt.savefig ()`, where the argument of the function is the name of the file that will contain the image:

```python
plt.savefig ("filename.png")
```

This feature **saves the graphic in `png` format and default resolution** in the same folder as the notebook.

This statement should be placed after all `plt` statements except` plt.show ()`. Failure to do so may result in a file being generated without the figure or missing some details (always check).

Copy and paste the contents of the previous code cell into the next cell, add the instructions needed to save a file named `curves.png`, and run it.

````python
%matplotlib inline

x = np.linspace (-2,2,101)
plt.xlabel ("x")
plt.ylabel ("f (x)")
y = x ** 2
plt.plot (x, y, label = "x ** 2") # Vectors x, y
y2 = x ** 4
plt.plot (x, y2, label = "x ** 4") # Vectors x, y2
# plt.savefig ("fig1.png")
plt.legend ()
plt.savefig ("fig2.png")
plt.show ()
# plt.savefig ("fig3.png")
````
